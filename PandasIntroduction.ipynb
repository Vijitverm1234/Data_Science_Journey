## Merging Data Frames

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['a','b','c','d']
})
df2=pd.DataFrame({
    'id':[3,4,7,8],
    'course':['a1','b1','c1','d1']
})
print(df1)
print(df2)
```

```python
inner_result=pd.merge(df1,df2,how='inner',on='id')
print(inner_result)
```

```python
inner_result=pd.merge(df1,df2,how='left',on='id')
print(inner_result)
```

```python
inner_result=pd.merge(df1,df2,how='right',on='id')
print(inner_result)
```

```python
inner_result=pd.merge(df1,df2,how='outer',on='id')
print(inner_result)
```

## Concatenating two DataFrames

```python
result=pd.concat([df1,df2],axis=0,ignore_index=True) # vertifically by default
print(result)
result=pd.concat([df1,df2],axis=0,keys=['df1','df2']) # Multi level indexing
print(result)
```
