```python
logo = np.load("/content/numpy logo.npy")

# Create a figure with two subplots
plt.figure(figsize=(10, 5))

# Display original logo
plt.subplot(121)
plt.imshow(logo)
plt.title("NumPy Logo")
plt.grid()

# Create and display dark logo
dark_logo = 1 - logo
plt.subplot(122)
plt.title("Dark Logo")
plt.imshow(dark_logo)
plt.grid()

# Show the figure
plt.show()
```
